```
install.packages("rstan")
```

```
library(rstan)
load("data_for_stan.RData")
print(params$stanfile)
options(mc.cores = parallel::detectCores())
print(data_list_england_wales)
```

```
risk_england_wales <- stan(params$stanfile, iter = params$iter, chains = 3,
  data = data_list_england_wales, seed = 1, control = list(adapt_delta = params$adapt_delta))
save(risk_england_wales,file="stan_output_england.Rdata")

```
